<h1>CWT Demo

First, we load a demo signal sampled at 200 Hz, so we can compute its CWT.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

demo_signal = np.loadtxt("demo_signal.csv")

fs = 200
time_axis = np.arange(0, demo_signal.shape[0])/fs

plt.figure(figsize=(15,2))
plt.plot(time_axis, demo_signal)
plt.title("Demo Signal")
plt.xlabel("Time [s]")
plt.show()
```

Now, let's compute its CWT transform using the complex morlet wavelet. Remember that the complex morlet wavelet has the formula given by:
$$\psi(k) = \frac{1}{\sqrt{\pi F_b}}e^{i2\pi F_ck} e^{-k^2/F_b} $$
In this implementation we fix $F_c=1$ and let you choose $F_b$ values and the desired frequency range, since different frequencies can be achieved by choosing appropiate scales. You can choose several $F_b$ values at once if that's what you want. The spectrograms obtained that way will be stacked along the channel dimension. For a given scale $s$, the scaled wavelet is computed as:
$$ \psi_s(k) = \frac{1}{\sqrt{s}}\psi\left(\frac{k}{s}\right) $$

The scales are automatically computed from the given frequency range, and they are equally spaced in a logarithmic axis. That is to say, the higher the frequency value, the greater the space between subsequent frequencies. This is recommended for wavelets transformations given their time-frequency trade-off.

One final detail: we let you decide whether you want to flatten the spectrogram. Usually, natural signals tend to have a power spectrum that decays like $1/f$, so lower frequencies will have more power in general. If you set $flattening$ to True, the wavelets will be multiplied by $f$ to avoid the $1/f$ behaviour, so the coefficients for higher frequencies will be amplified by $f$.

For this demo, let's use 32 scales between 3Hz and 40Hz. Since the demo signal has 1000 samples, we expect to have each spectrogram with shape (32, 1000). And say we want to obtain spectrograms for $F_b\in\{0.5,1.0,1.5,2.0\}$, so we expect to have 4 channels. We will also set flattening to True.

```python
# Parameters CWT
lower_freq = 3
upper_freq = 40
n_scales = 32
fb_array = np.array([0.5, 1.0, 1.5, 2.0])
```

For convenience, we compute the complex morlet wavelets in advance, in case you want to use the same wavelets in several operations. The wavelets are computed using only numpy operators, so this module can be run safely outside tensorflow.

```python
import cwt_ops

wavelets, frequencies = cwt_ops.complex_morlet_wavelets(
        fb_array=fb_array,
        fs=fs,
        lower_freq=lower_freq,
        upper_freq=upper_freq,
        n_scales=n_scales,
        flattening=True)
```

You can take a look at the frequency axis. The numbers are in Hz.

```python
print(frequencies)
```

And you can take a look at a wavelet. Let's show the 3Hz wavelet for $F_b=1.5$. 3Hz is associated with the biggest scale of the frequency range.

```python
which_fb = 2
which_scale = -1

real_bank, imag_bank = wavelets[which_fb]
single_real_wavelet = real_bank[0, :, 0, which_scale]
single_imag_wavelet = imag_bank[0, :, 0, which_scale]

time_axis = np.arange(0, single_real_wavelet.shape[0])/fs
time_axis = time_axis - np.mean(time_axis)

plt.figure(figsize=(10,4))
plt.plot(time_axis, single_real_wavelet)
plt.plot(time_axis, single_imag_wavelet)
plt.title("Complex Morlet Wavelet for " + str(frequencies[-1]) + " Hz")
plt.legend(["Real Part", "Imaginary Part"])
plt.xlabel("Time [s]")
plt.show()
```

Now, we will use the wavelets to compute the spectrograms of the demo signal. The implementation support a batch of signals, that is, a tensor of shape (batch size, signal length), so if you want to evaluate a single signal, you should enter a shape (1, signal length). For this demo, that can be achieved by simply using numpy.array([demo_signal]).

```python
import tensorflow as tf

tf.reset_default_graph()

spectrograms = cwt_ops.cwt_layer(
            inputs=np.array([demo_signal]),
            wavelets=wavelets)

with tf.Session() as sess:
    spectrograms_np = sess.run(spectrograms)
    
print("The result has shape", spectrograms_np.shape)

for j, fb in enumerate(fb_array):
    single_spectrogram = spectrograms_np[0, :, :, j]
    plt.figure(figsize=(15,2))
    plt.imshow(single_spectrogram, interpolation='none', cmap=cm.inferno, aspect='auto')
    plt.title("Spectrogram for Fb " + str(fb))
    plt.show()
```

You have the option of cropping the borders, to avoid border effects for the larger wavelets. For example, let's crop the first and the last second of the signal. That is, we will remove $1\times f_s=200$ samples at the beginning and at the end, so we expect spectrograms of shape (32, 600).

```python
tf.reset_default_graph()

# We crop 1 second at the beginning and at the end
border_crop = 1*fs

spectrograms = cwt_ops.cwt_layer(
            inputs=np.array([demo_signal]),
            wavelets=wavelets,
            border_crop=border_crop)

with tf.Session() as sess:
    spectrograms_np = sess.run(spectrograms)
    
print("The result has shape", spectrograms_np.shape)

for j, fb in enumerate(fb_array):
    single_spectrogram = spectrograms_np[0, :, :, j]
    plt.figure(figsize=(15,2))
    plt.imshow(single_spectrogram, interpolation='none', cmap=cm.inferno, aspect='auto')
    plt.title("Spectrogram for Fb " + str(fb))
    plt.show()
```

You can also compute the CWT with a stride in the time domain, to save computation. For example, we will compute the value of the CWT every 10 samples. If we maintain the cropping of the border in 1 second, we expect spectrograms of shape (32, 60).

```python
tf.reset_default_graph()

# We crop 1 second at the beginning and at the end
border_crop = 1*fs
# And we compute the CWT every 10 samples
stride = 10

spectrograms = cwt_ops.cwt_layer(
            inputs=np.array([demo_signal]),
            wavelets=wavelets,
            border_crop=border_crop,
            stride=stride)

with tf.Session() as sess:
    spectrograms_np = sess.run(spectrograms)
    
print("The result has shape", spectrograms_np.shape)

for j, fb in enumerate(fb_array):
    single_spectrogram = spectrograms_np[0, :, :, j]
    plt.figure(figsize=(15,2))
    plt.imshow(single_spectrogram, interpolation='none', cmap=cm.inferno, aspect='auto')
    plt.title("Spectrogram for Fb " + str(fb))
    plt.show()
```

Finally, you can see how the results change if we set flattening to False. Expect to have power concentrated at low frequencies, so variations at higher frequencies are harder to distinguish. We remove the cropping at the border and the stride so you can compare this results with the first one.

```python
wavelets, frequencies = cwt_ops.complex_morlet_wavelets(
        fb_array=fb_array,
        fs=fs,
        lower_freq=lower_freq,
        upper_freq=upper_freq,
        n_scales=n_scales,
        flattening=False)

tf.reset_default_graph()

spectrograms = cwt_ops.cwt_layer(
            inputs=np.array([demo_signal]),
            wavelets=wavelets)

with tf.Session() as sess:
    spectrograms_np = sess.run(spectrograms)
    
print("The result has shape", spectrograms_np.shape)

for j, fb in enumerate(fb_array):
    single_spectrogram = spectrograms_np[0, :, :, j]
    plt.figure(figsize=(15,2))
    plt.imshow(single_spectrogram, interpolation='none', cmap=cm.inferno, aspect='auto')
    plt.title("Spectrogram for Fb " + str(fb))
    plt.show()
```
